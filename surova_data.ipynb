```python
import pandas as pd

soubor_vstup = '{MarekVašíček}_surova_data.csv'
df = pd.read_csv(soubor_vstup, sep=';', encoding='utf-8-sig')
```

```python
# Základní informace o datech
print(f'Počet řádků a sloupců: {df.shape}')
print('\nNáhled dat (prvních 5 řádků):')
print(df.head())
print("\nDatové typy sloupců:")
print(df.dtypes)
```

```python
df['Cena bytu'] = df['Cena bytu'].str.replace('Kč', '')           # Smaže Kč
df['Cena bytu'] = df['Cena bytu'].str.replace('/měsíc', '')      # Smaže info o nájmu
df['Cena bytu'] = df['Cena bytu'].str.replace(' ', '')            # Smaže obyčejné mezery
df['Cena bytu'] = df['Cena bytu'].str.replace('\xa0', '')         # Smaže "neviditelné" mezery (častý problém webů)

df['Cena bytu'] = pd.to_numeric(df['Cena bytu'], errors='coerce')
```

```python
df['Rozmery bytu'] = df['Rozmery bytu'].str.replace('m²', '')
df['Rozmery bytu'] = df['Rozmery bytu'].str.replace(' ', '')
df['Rozmery bytu'] = df['Rozmery bytu'].str.replace('\xa0', '')

df['Rozmery bytu'] = pd.to_numeric(df['Rozmery bytu'], errors='coerce')
```

```python
print('Základní statistika datových typů po změně:')
print(df.dtypes)
```

```python
print('Základní statistika po vyčištění:')
print(f'Počet řádků a sloupců po očištění: {df.shape}')
```

```python
df.to_csv('{MarekVašíček}_zdrojova_data.csv', index=False, encoding='utf-8-sig', sep=';')
```
